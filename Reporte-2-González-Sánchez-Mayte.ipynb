Paises con más exportaciones por año

```python
exp_año = exportaciones[['year', 'origin', 'transport_mode', 'total_value']]
df = exp_año.groupby(['year', 'origin', 'transport_mode'])
top_export = df.count()['total_value'].sort_values(ascending=False).head(10)
top_export = top_export.reset_index()
sum_total = top_export['total_value'].sum()
top_export['Porciento'] = (top_export['total_value'] / sum_total) * 100
top_export['Por_acum'] = top_export.cumsum()['Porciento']
top_export
```

Paises con más importaciones por año

```python
imp_año = importaciones[['year', 'destination', 'transport_mode', 'total_value']]
df2 = imp_año.groupby(['year', 'destination', 'transport_mode'])
top_import = df2.count()['total_value'].sort_values(ascending=False).head(10)
top_import = top_import.reset_index()
sum_total2 = top_import['total_value'].sum()
top_import['Percent'] = (top_import['total_value'] / sum_total2) * 100
top_import['Por_acumu'] = top_import.cumsum()['Percent']
top_import
```

Productos más vendidos

```python
paises = exportaciones.groupby('origin').sum()
products_paises = exportaciones.groupby('origin')['product'].agg(pd.Series.mode)

paises['productos'] = products_paises
paises['porcentaje'] = (paises['total_value'] / paises['total_value'].sum()) * 100
paises['por_acum'] = paises.cumsum()['porcentaje']
paises.reset_index(inplace=True)
paises
```
